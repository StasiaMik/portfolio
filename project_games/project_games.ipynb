Рейтинг ESRB (Entertainment Software Rating Board) не присваивается всем играм. Рейтинг ESRB обычно присваивается только тем играм, которые распространяются в Северной Америке и выпускаются через официальные каналы. Вот несколько причин, по которым игра может не иметь рейтинга ESRB:
- Малобюджетные и независимые игры: Многие независимые разработчики или небольшие студии могут не иметь ресурсов для получения рейтинга ESRB, так как процесс получения рейтинга требует оплаты и времени.
- Региональные различия: Игры, выпущенные в других регионах, могут иметь рейтинг других организаций. Например, PEGI используется в Европе, CERO в Японии, а ACB в Австралии.
- Онлайн-игры и приложения: Многие онлайн-игры, мобильные приложения и браузерные игры могут не иметь рейтинга ESRB, особенно если они распространяются через неофициальные или независимые платформы.
- Старые игры: Некоторые старые игры, выпущенные до введения рейтинговой системы ESRB в 1994 году, могут не иметь рейтинга.
- Пользовательский контент: Игры с большим количеством пользовательского контента, такие как платформы для создания игр или игровые песочницы, могут не иметь фиксированного рейтинга, так как содержание может значительно варьироваться.

Для анализа данных, связанных с рейтингом ESRB, важно учитывать, что не все записи могут содержать эти данные, и это нужно обрабатывать соответствующим образом в нашем анализе.

```python
#в рейтинге ESRB заменим пропуски на заглушку without
df['rating'] = df['rating'].fillna('without')
```

 **Построим распределение по годам для топ-6 платформ**

```python
#введём отдельный df с платформами с самой высокой доходностью
temp = df.copy() 
top_platform_df = temp.query('platform in ["PS2", "X360", "PS3", "Wii", "DS", "PS"]')

#зададим порядок топ-6 платформ 
ordered_platforms = list(temp.groupby('platform')['total_sales'].sum().sort_values(ascending=False).index[:6])

#построим сводную таблицу и графики для каждой платформы
for platform in ordered_platforms:
    platform_df = top_platform_df[top_platform_df['platform'] == platform]
    pivot_df = platform_df.pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')
    print(f"Платформа: {platform}")
    display(pivot_df)
    
    pivot_df.plot(kind='bar', color='lightcoral', figsize=(8, 4), legend=False)
    plt.title(f'Продажи по годам для {platform}')
    plt.xlabel('Год выпуска')
    plt.ylabel('Сумма продаж')
    plt.show()
```

### Шаг 4. Составим портрет пользователя каждого региона

Определим для пользователя каждого региона (NA, EU, JP):  Самые популярные платформы (топ-5)

```python
#напишим функцию для группировки 
def calculate_sales_percentage(df, region):
    platform_sales = df.groupby(by='platform').agg({region: 'sum'}).sort_values(by=region, ascending=False)
    platform_sales[f'{region}_per'] = (platform_sales[region] / platform_sales[region].sum()) * 100
    return platform_sales

#функция для построения столбцов
def plot_sales_percentage(platform_sales, region, color='lightcoral'):
    plt.figure(figsize=(8, 4))
    bars = plt.bar(platform_sales.index, platform_sales[region], color=color)

    #добавляем проценты на столбцы
    for bar, pct in zip(bars, platform_sales[f'{region}_per']):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height + 1, f'{pct:.1f}%', ha='center', va='bottom')

    plt.title(f'Самые популярные платформы в регионе ({region})')
    plt.xlabel('Платформа')
    plt.ylabel('Продажи')
    plt.xticks(rotation=45)
    plt.ylim(0, platform_sales[region].max() + 30)
    plt.show()
```

```python
jp_platform_sales = calculate_sales_percentage(actual_df, 'jp_sales')

print(jp_platform_sales.head(5))

plot_sales_percentage(jp_platform_sales, 'jp_sales')
```

```python
#функция для построения круговой диаграммы с группировкой "Другие"
def plot_pie_chart(data, column, region, region_name, title):
    #группируем продажи по жанрам/платформам и сортируем
    sales = data.groupby(column)[region].sum().sort_values(ascending=False)   
    #определяем топ-5
    top_items = sales.head(5)
    #остальные продажи объединяем в "Другие"
    other_sales = sales[5:].sum()
    #добавляем в список
    sales_data = top_items.append(pd.Series(other_sales, index=['Другие']))
    
    #визуализация
    plt.figure(figsize=(7, 7))
    plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
    plt.title(f'{title} в {region_name}')
    plt.show()

#анализ по жанрам
plot_pie_chart(actual_df, 'genre', 'jp_sales', 'Япония', 'Популярные жанры')
plot_pie_chart(actual_df, 'genre', 'eu_sales', 'Европа', 'Популярные жанры')
plot_pie_chart(actual_df, 'genre', 'na_sales', 'Северная Америка', 'Популярные жанры')

#анализ по платформам
plot_pie_chart(actual_df, 'platform', 'jp_sales', 'Япония', 'Популярные платформы')
plot_pie_chart(actual_df, 'platform', 'eu_sales', 'Европа', 'Популярные платформы')
plot_pie_chart(actual_df, 'platform', 'na_sales', 'Северная Америка', 'Популярные платформы')
```

### Шаг 5. Проверим гипотезы:

#### Гипотеза №1
Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

    Нулевая гипотеза (H₀): Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Это означает, что в рамках данной гипотезы мы предполагаем, что нет статистически значимых различий между средними пользовательскими рейтингами игр на платформах Xbox One и PC. Формулировка нулевой гипотезы подразумевает, что любые наблюдаемые различия могут быть объяснены случайными факторами.

    Альтернативная гипотеза (H₁): Средние пользовательские рейтинги платформ Xbox One и PC различаются.

Альтернативная гипотеза предполагает, что различия в средних пользовательских рейтингах между платформами Xbox One и PC не являются случайными и могут быть объяснены наличием реальных различий между платформами.

Так как наши выборки независимы между собой будем использовать t-критерий Стьюдента

```python
xbox_one = actual_df[(actual_df['platform'] == 'XOne') & (actual_df['user_score'].notnull())]['user_score']
pc = actual_df[(actual_df['platform'] == 'PC') & (actual_df['user_score'].notnull())]['user_score']

alpha = 0.05
results = st.ttest_ind(xbox_one, pc)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
print('Средняя оценка пользователей игр для платформы Xbox One  - {:.2f}'.format(xbox_one.mean()))
print('Средняя оценка пользователей игр для платформы PC - {:.2f}'.format(pc.mean()))
```

Результат: нет достаточных данных для того, чтобы утверждать, что средние пользовательские рейтинги платформ Xbox One и PC различаются на статистически значимом уровне.

#### Гипотеза №2 
Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

    Нулевая гипотеза (H₀): Средние пользовательские рейтинги жанров Action и Sports одинаковые.

Эта гипотеза предполагает, что любые наблюдаемые различия в средних пользовательских рейтингах между жанрами Action и Sports могут быть объяснены случайными факторами или шумом в данных. Формулировка H₀ указывает на отсутствие статистически значимого различия между жанрами.

    Альтернативная гипотеза (H₁): Средние пользовательские рейтинги жанров Action и Sports различаются.

В рамках этой гипотезы предполагается, что существует статистически значимое различие в средних пользовательских рейтингах между жанрами Action и Sports. Альтернативная гипотеза указывает на наличие реальных различий в рейтингах, которые можно обнаружить и подтвердить с помощью статистического анализа данных.

Здесь так же будем использовать t-критерий Стьюдента

```python
action_ratings = actual_df[(actual_df['genre'] == 'Action') & (actual_df['user_score'].notnull())]['user_score']
sports_ratings = actual_df[(actual_df['genre'] == 'Sports') & (actual_df['user_score'].notnull())]['user_score']

alpha = 0.05
results = st.ttest_ind(action_ratings, sports_ratings)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Вывод: Средние пользовательские рейтинги жанров Action и Sports различаются, что означает, что у нас есть статистически значимые доказательства того, что средние пользовательские рейтинги для жанров Action и Sports различаются, на имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.
